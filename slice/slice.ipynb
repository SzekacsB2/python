# slice of a heightmap

extracts a line segment from a 2D image between two specified points

the function samples *N* equally spaced points along the line, where *N* is the integer part of the Euclidean distance between `pt1` and `pt2`

### parameters

- **image** (`np.ndarray`): a 2D array representing the grayscale image or heightmap
- **pt1** (`tuple`): the starting point of the line segment
- **pt2** (`tuple`): the ending point of the line segment, *not included* in the sampling

### returns

- the list of sampled distances
- the array of corresponding intensity values

```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
def slice(image, pt1, pt2):
    dist = np.sqrt((pt1[0]-pt2[0])**2 + (pt1[1]-pt2[1])**2)
    N = int(dist)
    rows = [int(pt1[0] + (pt2[0]-pt1[0])/N*i) for i in range(N)]
    cols = [int(pt1[1] + (pt2[1]-pt1[1])/N*i) for i in range(N)]
    dists = [dist/N*i for i in range(N)]
    return dists, image[rows, cols]
```

---

```python
def plot(dists, heights):
    plt.plot(dists, heights)

    plt.figure()
    plt.pcolormesh(image)
    plt.gca().set_aspect(1)
    plt.plot([pt1[1],pt2[1]],[pt1[0],pt2[0]],color='white')
```

#### e.g. 1:

```python
image = np.array(Image.open('./map1.png').convert('L'))
pt1 = (672, 23)
pt2 = (137, 716)

dists, heights = slice(image, pt1, pt2)
plot(dists, heights)
```
